4 a.	Write a python program to find out top 15 jobs. Make a bar graph of sales as well. 

```python
# Grouping the data by groupby() function. 'job_title' and calculating the sum of salaries for each job title
job_sales = df.groupby('job_title')['salary_in_usd'].sum().sort_values(ascending=False)

# Selecting the top 15 jobs with the highest total salaries
top_15_jobs = job_sales.head(15)
top_15_jobs # Printing top 15 jobs 
```

```python
# Setting up the figure size for the plot
plt.figure(figsize=(12, 8))

# Creating a bar plot of the top 15 jobs and their total salaries
top_15_jobs.plot(kind='bar', color='blue')

# Adding title and labels to the plot
plt.title('Top 15 Jobs by Salary (Sales)')
plt.xlabel('Job Title')
plt.ylabel('Total Salary (Sales) in USD')
plt.xticks(rotation=90) # text of x lable are tilted 
plt.show()
```

4 b.	Which job has the highest salaries? Illustrate with bar graph.

```python
job_salaries = df.groupby('job_title')['salary_in_usd'].mean() # mean calculate average salary for each group

# job title with the maximum value (highest average salary) in the job_salaries column
highest_salary_job = job_salaries.idxmax()

# Creating bar graph of top job title.
plt.figure(figsize=(8, 6))
plt.bar(highest_salary_job, job_salaries[highest_salary_job],color='blue')
plt.xlabel('Job Title')
plt.ylabel('Mean Salary (USD)')
plt.title('Job with Highest Salary')
plt.show()
```

4 c.	Write a python program to find out salaries based on experience level. Illustrate it through bar graph.

```python
# Grouping the data by 'experience_level' and calculating the mean salary for each experience level 
mean_salary_by_experience = df.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False)
mean_salary_by_experience
```

```python
# Setting up the figure size for the plot
plt.figure(figsize=(5,6))

# Creating a bar plot of the mean salary for each experience level
mean_salary_by_experience.plot(kind='bar',color='blue')

# Adding title and labels to the plot
plt.title('Mean Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Mean Salary (in USD)')

# Rotating x-axis labels for better understanding
plt.xticks(rotation=90,)

# Adjusting layout to prevent labels from being cut off
plt.tight_layout()

# Displaying the plot
plt.show()
```

```python
def plot_histogram_and_boxplot(data, variable_name):  
    # Plot histogram
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)  # Create a subplot for histogram
    plt.hist(data, bins=20, color='blue', edgecolor='black')
    plt.title('Histogram of ' + variable_name)
    plt.xlabel(variable_name)
    plt.ylabel('Frequency')

    # Plot box plot
    plt.subplot(1, 2, 2)  # Create a subplot for box plot
    plt.boxplot(data, vert=False)
    plt.title('Box Plot of ' + variable_name)
    plt.xlabel(variable_name)
    plt.yticks([])  # Hide y-axis ticks
    plt.tight_layout()
    plt.show()
#
variable_name = 'salary_in_usd'
plot_histogram_and_boxplot(df[variable_name], variable_name)
```
